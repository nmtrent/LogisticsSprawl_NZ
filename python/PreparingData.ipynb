## Read in **A files**

Some conversion steps are first executed outside of the notebook. These manual steps include:
* Create copies of raw files by appending _copy (do this in terminal)

* Double check date in Year field is the same as year in the file name

* Save copy as xls overwriting the current file

### Tasman region issue
I  realised later (in April) that for some reason then Tasman Region files have duplicates of all areas. So instead of 33 areas, there are 66. I manually deleted the duplicates for the copiesForPython and then reran the code to create the GroupA file. 


```python
from pandas import DataFrame, read_csv
import matplotlib.pyplot as plt
import pandas as pd 

def readConvert(file,region,year): 
    df = pd.read_excel(file,header=None)
    #delete the empty second column
    df=df.drop(1,axis=1)
    #delete the last row which contains details about the download
    df=df.drop(len(df)-1,axis=0)
    #rename columns
    cols=["Area","TotInd_GeogUnits","TotInd_EmpCo","F_GeogUnits","F_EmpCo","G_GeogUnits","G_EmpCo","I461_GeogUnits","I461_EmpCo","I471_GeogUnits","I471_EmpCo","I481_GeogUnits","I481_EmpCo","I51_GeogUnits","I51_EmpCo","I53_GeogUnits","I53_EmpCo"]
    df.columns=cols
    #add a parent column
    recur = [region]*(len(df)-2)
    front =["TOTAL","NewZealand"]
    parentArea =  front+ recur
    df["ParentArea"]=parentArea
    #add year column
    yearColumn=[year]*len(df)
    df["Year"]=yearColumn

    #convert all columns to numeric
    df['TotInd_GeogUnits']=pd.to_numeric(df['TotInd_GeogUnits'],errors='coerce')
    df['TotInd_EmpCo']=pd.to_numeric(df['TotInd_EmpCo'],errors='coerce')
    df['F_GeogUnits']=pd.to_numeric(df['F_GeogUnits'],errors='coerce')
    df['F_EmpCo']=pd.to_numeric(df['F_EmpCo'],errors='coerce')
    df['G_GeogUnits']=pd.to_numeric(df['G_GeogUnits'],errors='coerce')
    df['G_EmpCo']=pd.to_numeric(df['G_EmpCo'],errors='coerce')
    df['I461_GeogUnits']=pd.to_numeric(df['I461_GeogUnits'],errors='coerce')
    df['I461_EmpCo']=pd.to_numeric(df['I461_EmpCo'],errors='coerce')
    df['I471_GeogUnits']=pd.to_numeric(df['I471_GeogUnits'],errors='coerce')
    df['I471_EmpCo']=pd.to_numeric(df['I471_EmpCo'],errors='coerce')
    df['I481_GeogUnits']=pd.to_numeric(df['I481_GeogUnits'],errors='coerce')
    df['I481_EmpCo']=pd.to_numeric(df['I481_EmpCo'],errors='coerce')
    df['I51_GeogUnits']=pd.to_numeric(df['I51_GeogUnits'],errors='coerce')
    df['I51_EmpCo']=pd.to_numeric(df['I51_EmpCo'],errors='coerce')
    df['I53_GeogUnits']=pd.to_numeric(df['I53_GeogUnits'],errors='coerce')
    df['I53_EmpCo']=pd.to_numeric(df['I53_EmpCo'],errors='coerce')
    print('Complete :'+region+" "+year+str(df.shape))
    return(df)


regionPathMaster=['AreaOutsideRegion/A-GeographicUnitsEmployeeCount/CopiesForPython/A_AreaOutsideRegion_',
                  'AucklandRegion/A-GeographicUnitsEmployeeCount/CopiesForPython/A_AucklandRegion_',
                 'BayOfPlentyRegion/A-GeographicUnitsEmployeeCount/CopiesForPython/A_BayOfPlentyRegion_',
                  'CanterburyRegion/A-GeographicUnitsEmployeeCount/CopiesForPython/A_CanterburyRegion_',
                  'GisborneRegion/A-GeographicUnitsEmployeeCount/CopiesForPython/A_GisborneRegion_',
                  'HawkesBayRegion/A-GeographicUnitsEmployeeCount/CopiesForPython/A_HawkesBayRegion_',
                  'Manawatu-WanganuiRegion/A-GeographicUnitsEmployeeCount/CopiesForPython/A_Manawatu-WanganuiRegion_',
                  'MarlboroughRegion/A-GeographicUnitsEmployeeCount/CopiesForPython/A_MarlboroughRegion_',
                  'NelsonRegion/A-GeographicUnitsEmployeeCount/CopiesForPython/A_NelsonRegion_',
                  'NorthlandRegion/A-GeographicUnitsEmployeeCount/CopiesForPython/A_NorthlandRegion_',
                  'OtagoRegion/A-GeographicUnitsEmployeeCount/CopiesForPython/A_OtagoRegion_',
                  'SouthlandRegion/A-GeographicUnitsEmployeeCount/CopiesForPython/A_SouthlandRegion_',
                  'TaranakiRegion/A-GeographicUnitsEmployeeCount/CopiesForPython/A_TaranakiRegion_',
                  'TasmanRegion/A-GeographicUnitsEmployeeCount/CopiesForPython/A_TasmanRegion_',
                  'WaikatoRegion/A-GeographicUnitsEmployeeCount/CopiesForPython/A_WaikatoRegion_',
                  'WellingtonRegion/A-GeographicUnitsEmployeeCount/CopiesForPython/A_WellingtonRegion_',
                  'WestCoastRegion/A-GeographicUnitsEmployeeCount/CopiesForPython/A_WestCoastRegion_']
regionMaster=['AreaOutsideRegion','AucklandRegion','BayOfPlentyRegion','CanterburyRegion','GisborneRegion','HawkesBayRegion',
             'Manawatu-WanganuiRegion','MarlboroughRegion','NelsonRegion','NorthlandRegion','OtagoRegion','SouthlandRegion',
              'TaranakiRegion','TasmanRegion','WaikatoRegion','WellingtonRegion','WestCoastRegion']
years=['2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020']

completeData=readConvert('AreaOutsideRegion/A-GeographicUnitsEmployeeCount/CopiesForPython/A_AreaOutsideRegion_2000_copy.xls',"AreaOutsideRegion","2000")
for area in range(len(regionPathMaster)): 
    for year in range(len(years)):
        filename=regionPathMaster[area]+years[year]+"_copy.xls"
        if filename !='AreaOutsideRegion/A-GeographicUnitsEmployeeCount/CopiesForPython/A_AreaOutsideRegion_2000_copy.xls':
            completeData=completeData.append(readConvert(filename,regionMaster[area],years[year]))


completeData.to_csv("CompleteSet_GroupA.csv")
print('Processing complete')
```

```python
from pandas import DataFrame, read_csv
import matplotlib.pyplot as plt
import pandas as pd 

regionMaster=['AreaOutsideRegion','AucklandRegion','BayOfPlentyRegion','CanterburyRegion','GisborneRegion','HawkesBayRegion',
             'Manawatu-WanganuiRegion','MarlboroughRegion','NelsonRegion','NorthlandRegion','OtagoRegion','SouthlandRegion',
              'TaranakiRegion','TasmanRegion','WaikatoRegion','WellingtonRegion','WestCoastRegion']
folderMaster=['D-EmploymentIncome','E-PopulationEthnicGroups','F-WorkLabourStatus']



def readConvertD2001(file,region): 
    df = pd.read_excel(file,header=None)
    #delete the empty second column
    df=df.drop(1,axis=1)
    #add column names
    cols=["Area",
          "loss_TotInd","loss_Wholesale","loss_Retail","loss_TransStor",
          "zero_TotInd","zero_Wholesale","zero_Retail","zero_TransStor",
          "less5k_TotInd","less5k_Wholesale","less5k_Retail","less5k_TransStor",
          "bet5k10k_TotInd","bet5k10k_Wholesale","bet5k10k_Retail","bet5k10k_TransStor",
          "bet10k15k_TotInd","bet10k15k_Wholesale","bet10k15k_Retail","bet10k15k_TransStor",
          "bet15k20k_TotInd","bet15k20k_Wholesale","bet15k20k_Retail","bet15k20k_TransStor",
          "bet20k25k_TotInd","bet20k25k_Wholesale","bet20k25k_Retail","bet20k25k_TransStor",
          "bet25k30k_TotInd","bet25k30k_Wholesale","bet25k30k_Retail","bet25k30k_TransStor",
          "bet30k40k_TotInd","bet30k40k_Wholesale","bet30k40k_Retail","bet30k40k_TransStor",
          "bet40k50k_TotInd","bet40k50k_Wholesale","bet40k50k_Retail","bet40k50k_TransStor",
          "bet50k70k_TotInd","bet50k70k_Wholesale","bet50k70k_Retail","bet50k70k_TransStor",
          "bet70k100k_TotInd","bet70k100k_Wholesale","bet70k100k_Retail","bet70k100k_TransStor",
          "greater100k_TotInd","greater100k_Wholesale","greater100k_Retail","greater100k_TransStor",
          "notStated_TotInd","notStated_Wholesale","notStated_Retail","notStated_TransStor"]
    df.columns=cols
    #convert columns to numbers
    for i in range(1,57):
        df.iloc[:,i]=pd.to_numeric(df.iloc[:,i],errors='coerce')
    
    #add first two areas
    if region == "MarlboroughRegion":
        df.iloc[0,0]=region
        #add a parent column
        recur = [region]*(len(df)-1)
        front =["NewZealand"]
        parentArea =  front+ recur
        df["ParentArea"]=parentArea
    else:
        df.iloc[0,0]="TotalNZ"
        df.iloc[1,0]=region
        #add a parent column
        recur = [region]*(len(df)-2)
        front =["TOTAL","NewZealand"]
        parentArea =  front+ recur
        df["ParentArea"]=parentArea      
    #add year column
    yearColumn=["2001"]*len(df)
    df["Year"]=yearColumn
    print('Complete :'+region+" D "+str(df.shape))
    return(df)

def readConvertE2001(file,region): 
    formatB=["AreaOutsideRegion","CanterburyRegion","MarlboroughRegion","NelsonRegion","OtagoRegion","SouthlandRegion","TasmanRegion","WestCoastRegion"]
    df = pd.read_excel(file,header=None)
    #delete the empty second column
    df=df.drop(1,axis=1)
    if region in formatB:
        #add column names
        cols=["Area","TotUsuallyResident","European","Maori","Pacific","Asia","Other","TotalSpec1orMore"]
        df.columns=cols
    else:
        #add column names
        cols=["Area","European","Maori","Pacific","Asia","Other","TotalSpec1orMore","TotUsuallyResident"]
        df.columns=cols
    #convert columns to numbers
    for i in range(1,8):
        df.iloc[:,i]=pd.to_numeric(df.iloc[:,i],errors='coerce')

    if region=="SouthlandRegion":
        df.iloc[0,0]=region
        #add a parent column
        recur = [region]*(len(df)-1)
        front =["NewZealand"]
        parentArea =  front+ recur
        df["ParentArea"]=parentArea 
    else:
        #add first two areas
        df.iloc[0,0]="TotalNZ"
        df.iloc[1,0]=region
        #add a parent column
        recur = [region]*(len(df)-2)
        front =["TOTAL","NewZealand"]
        parentArea =  front+ recur
        df["ParentArea"]=parentArea 
    #add year column
    yearColumn=["2001"]*len(df)
    df["Year"]=yearColumn
    print('Complete :'+region+" E "+str(df.shape))
    return(df)

def readConvertF2001(file,region): 
    df = pd.read_excel(file,header=None)
    #delete the empty second column
    df=df.drop(1,axis=1)
    #add column names
    cols=["Area","TotPeople","EmployeeFull","SelfEmpFull","EmployerFull","FamUnpaidFull","UnstatedFull","TotFull",
         "EmployeePart","SelfEmpPart","EmployerPart","FamUnpaidPart","UnstatedPart","TotPart",
          "TotPaid","TotSelfEmp","TotEmployer","TotUnpaidFam","TotUnstated","TotEmployed","Unemp","TotLabForce","NotLabForce","Unident"]
    df.columns=cols
    #add first two areas
    df.iloc[0,0]="TotalNZ"
    df.iloc[1,0]=region
    #convert columns to numbers
    for i in range(1,24):
        df.iloc[:,i]=pd.to_numeric(df.iloc[:,i],errors='coerce')
    #add a parent column
    recur = [region]*(len(df)-2)
    front =["TOTAL","NewZealand"]
    parentArea =  front+ recur
    df["ParentArea"]=parentArea 
    #add year column
    yearColumn=["2001"]*len(df)
    df["Year"]=yearColumn
    print('Complete :'+region+" F "+str(df.shape))
    return(df)

for folder in folderMaster:  
    for region in regionMaster: 
        if folder=='D-EmploymentIncome':
            filename=region+'/'+folder+'/CopiesForPython/D_'+region+'_2001_copy.xls'
            writefile="CompleteSet_2001_GroupD.csv"
            if filename=='AreaOutsideRegion/D-EmploymentIncome/CopiesForPython/D_AreaOutsideRegion_2001_copy.xls':
                completeData=readConvertD2001(filename,region)
            else:
                completeData=completeData.append(readConvertD2001(filename,region)) 
        elif folder == 'E-PopulationEthnicGroups':
            filename=region+'/'+folder+'/CopiesForPython/E_'+region+'_2001_copy.xls'
            writefile="CompleteSet_2001_GroupE.csv"
            if filename=='AreaOutsideRegion/E-PopulationEthnicGroups/CopiesForPython/E_AreaOutsideRegion_2001_copy.xls':
                completeData=readConvertE2001(filename,region)
            else:
                completeData=completeData.append(readConvertE2001(filename,region)) 
        else:
            filename=region+'/'+folder+'/CopiesForPython/F_'+region+'_2001_copy.xls'
            writefile="CompleteSet_2001_GroupF.csv"
            if filename=='AreaOutsideRegion/F-WorkLabourStatus/CopiesForPython/F_AreaOutsideRegion_2001_copy.xls':
                completeData=readConvertF2001(filename,region)
            else:
                completeData=completeData.append(readConvertF2001(filename,region)) 
    completeData.to_csv(writefile)
    print("Wrote out "+writefile)
```

```python
import pandas as pd 
from pandas import DataFrame, read_csv

regionMaster=['AreaOutsideRegion','AucklandRegion','BayOfPlentyRegion','CanterburyRegion','GisborneRegion','HawkesBayRegion',
             'Manawatu-WanganuiRegion','MarlboroughRegion','NelsonRegion','NorthlandRegion','OtagoRegion','SouthlandRegion',
              'TaranakiRegion','TasmanRegion','WaikatoRegion','WellingtonRegion','WestCoastRegion']

years=['2013','2018']

def readConvertG(file,region,year): 
    df = pd.read_excel(file,header=None)
    #delete the empty second column
    df=df.drop(1,axis=1)
    #delete last three rows of annotations
    df=df.drop(df.index[len(df)-3:len(df)],axis=0)
    #add column names
    cols=["Area",
          "TotHousehold","MedInc","less20k","bet20k_30k","bet30k_50k",
          "bet50k_70k","bet70k_100k","bet100k_150k","greater150k",
          "totStated","totNotStated"]
    df.columns=cols
    #convert columns to numbers
    for i in range(1,11):
        df.iloc[:,i]=pd.to_numeric(df.iloc[:,i],errors='coerce')
    #add a parent column
    recur = [region]*(len(df)-2)
    front =["TOTAL","NewZealand"]
    parentArea =  front+ recur
    df["ParentArea"]=parentArea 
    #add year column
    yearColumn=[year]*len(df)
    df["Year"]=yearColumn
    print('Complete :'+region+" G "+year+str(df.shape))
    return(df)


for year in years:
    for region in regionMaster: 
        filename=region+'/G-HouseholdIncome/CopiesForPython/G_'+region+'_'+year+'_copy.xls'
        writefile="CompleteSet_GroupG.csv"
        if filename=='AreaOutsideRegion/G-HouseholdIncome/CopiesForPython/G_AreaOutsideRegion_2013_copy.xls':
            completeData=readConvertG(filename,region,year)
        else:
            completeData=completeData.append(readConvertG(filename,region,year)) 

completeData.to_csv(writefile)
print("Wrote out "+writefile)

```
